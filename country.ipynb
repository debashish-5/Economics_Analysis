## Level - Intermediate🟡 

### Line plot
Qn.How does average GDP change over the Years ?

```python
avg_gdp = df.groupby("Year")["GDP (Billion USD)"].mean()
plt.plot(avg_gdp.index,avg_gdp.values,"b^:")
plt.title("Average GDP over Years")
plt.xlabel("Years")
plt.ylabel("GDP (Billion USD)")
plt.grid(True)
plt.tight_layout()
```

```python
avg_gdp
```

### Horizontal Bar Plot
Qn.Which countries have more internet users on average ?

```python
avg_internet = df.groupby("Country")["Internet Users (%)"].mean()
plt.barh(avg_internet.index,avg_internet.values,color="green")
plt.xlabel("Internet Users")
plt.ylabel("Country")
plt.title("Avg Internet Users by Countries")
```

### Bar plot
Qn.HOW does unemployeement rate vary by countries ?

```python
avg_unemp = df.groupby("Country")["Unemployment Rate (%)"].mean()
plt.bar(avg_unemp.index,avg_unemp.values,color="black",alpha=0.8)
plt.xticks(rotation=45)
plt.xlabel("Year")
plt.ylabel("Unemployment Rate (%)")
plt.title("Unemployment Rate (%) Vs Country")
```

### Pie chart
Qn.What's distribution of data across continents ?


```python
continent_count=df["Continent"].value_counts()
plt.pie(continent_count.values,
        labels=continent_count.index,
        startangle=140,
        autopct="%1.1f%%",
        wedgeprops={"edgecolor":"black","linewidth":2,"linestyle":"--"}
       )
plt.title("Data Distribution by Continent")
plt.axis("equal")
plt.tight_layout()
```

### Stack Plot
Qn.How is average GDP distributed across countries by continent ?

```python
grouped = df.groupby(["Country","Continent"])["GDP (Billion USD)"].mean().unstack(fill_value=0)
grouped.plot(kind='bar',stacked=True,figsize=(10,6),colormap="Set2")
plt.title("Stacked Bar: Avg GDP by Country & Continent")
plt.xlabel("Country")
plt.ylabel("GDP(Billion USD)")
plt.xticks(rotation=45)
plt.legend(title="Continent")
plt.tight_layout()
```

### Scatter plot
Qn.Is there a relationship between GDP and Internet Users ?(minimum / maximum)

```python
h=df.groupby("GDP (Billion USD)")["Internet Users (%)"].max()
l=df.groupby("GDP (Billion USD)")["Internet Users (%)"].min()
plt.scatter(h.index,h.values,label="Max")
plt.scatter(l.index,l.values,label="Min")
plt.legend()
```

### Histogram
Qn.How is the population distribution overall ?

```python

plt.hist(df['Population'],bins=30,edgecolor="white",color='purple',label="Population")
plt.title("Population Distribution")
plt.xlabel("Population")
plt.ylabel("Frequency")
plt.tight_layout()
plt.legend()
```

### Subplot
Qn.Show GDP and Internet Users in Two subplots ?

```python
avg_gdp = df.groupby("Country")["GDP (Billion USD)"].mean()
avg_net = df.groupby("Country")["Internet Users (%)"].mean()

plt.subplot(1,2,1)
plt.bar(avg_gdp.index,avg_gdp.values)
plt.title("Average GDP")
plt.xticks(rotation=45)
plt.tight_layout()


plt.subplot(1,2,2)
plt.bar(avg_net.index,avg_net.values,color="green")
plt.xticks(rotation=45)
plt.title("Internet Users (%)")
plt.tight_layout()
```

### Relplot
Qn.Which countries have higher GDP and lower unemployment ?

```python
plot=sns.relplot(x="GDP (Billion USD)",
                y="Unemployment Rate (%)",
                hue="Country",
                size="Internet Users (%)",
                sizes=(40,200),
                alpha=0.6,
                palette="muted",
                height=6,
                aspect=1.5,
                data=df
               )
plot.fig.suptitle("GDP Vs Unemployment Rate with Internet Usages as Size",fontsize=14)
plt.tight_layout()
                
```

### Heatmap
Qn.How are GDP, Unemployment Rate, Internet Users %, and Population correlated across all countries and years in the dataset?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Select only numerical columns for correlation
numeric_cols = ['Population', 'GDP (Billion USD)', 'Unemployment Rate (%)', 'Internet Users (%)']
corr_matrix = df[numeric_cols].corr()

# Plotting the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='mako', linewidths=0.5)
plt.title('Correlation Heatmap of Economic & Demographic Indicators')
plt.show()
```
